```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (1).csv')
df.head()
df.info()
df.describe()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

auto_count = (df['am'] == 0).sum()
manual_count = (df['am'] == 1).sum()
labels = ['Automatic', 'Manual']
counts = [auto_count, manual_count]

plt.bar(labels, counts, color=['orange', 'blue'])

plt.title("Transmission Types")
plt.xlabel("Transmission Type")
plt.ylabel("Number of Cars")
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.scatter(df['wt'], df['mpg'], color='red')
plt.title("Car Weight vs MPG")
plt.xlabel("Weight (1000 lbs)")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.hist(df['mpg'], bins=8, color='red')
plt.title("Distribution of MPG")
plt.xlabel("Miles per Gallon")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.boxplot(df['mpg'])
plt.title("Boxplot of MPG")
plt.xlabel("MPG")
plt.ylabel("Miles per Gallon")
plt.show()
```
