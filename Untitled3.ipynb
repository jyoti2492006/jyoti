```python
import pandas as pd
df = pd.read_csv('/content/titanic.csv')
df.head()
df.info()
df.describe()
df.isnull().sum()
df.dropna()
df.fillna(0)
```

```python
print(df.isnull().sum)
```

```python
missing_survived = df['Survived'].isnull().sum()
missing_age = df['Age'].isnull().sum()

print(f"Missing values in 'Survived': {missing_survived}")
print(f"Missing values in 'Age': {missing_age}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
Pclass_counts = df['Pclass'].value_counts()
plt.bar(Pclass_counts.index, Pclass_counts.values)
plt.title('pclass distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```
